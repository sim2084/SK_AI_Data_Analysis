# 통계 기초
Numpy 통계는 정식 통계분석 도구라기보다, 데이터를 빠르게 요약하고 분포를 감 잡기 위한 기초 연산이다.

```python
import numpy as np
```

## 중심 경향 축도(Central Tendency)
데이터의 대표적인 중심 값을 나타내는 통꼐 지표이다.
대표적으로 다음 세가지를 많이 사용한다.
- 평균(mean) : 모든 값의 합을 데이터의 개수로 나눈 값
- 중앙값(median) : 데이터를 정렬했을때 가운데 위치한 값
- 최빈값(mode) : 가장 많이 등장하는 값

```python
# 학생들의 시험점수
scores = np.array([85, 90, 92, 88, 90, 75, 88, 95, 92, 85, 80, 95, 100, 92, 85])
```

```python
# 평균(Mean) : 데이터의 전체적인 중심을 나타내는 가장 기본적인 통계값
print(f'합계 : {np.sum(scores)}')
print(f'데이터 개수 : {scores.size}')
print(f'평균 : {np.mean(scores)}')
```

```python
# 중앙값(median) : 데이터를 정렬했을때 가운데 위치한 값
# 홀수 -> 가운데 1개 값
# 짝수 -> 가운데 2개 값의 평균
# 이상치(outlier)의 영향을 평균보다 덜 받는 특징
print(sorted(scores))
print(f'중앙값 : {np.median(scores)}')
```

```python
# 최빈값(mode) : 데이터에서 가장 많이 등장하는 값
print(f'중복 제거한 값 : {np.unique(scores)}')
values, counts = np.unique(scores,return_counts=True)
print(values,counts)

max_count = np.max(counts)
# 최대 빈도수와 같은 위치의 값만 선택
vals = values[counts==max_count]
print(f'최빈값 : {vals}, 빈도 수 : {max_count}')
```

## 산포 측도
데이터가 평균을 기준으로 얼마나 퍼졌 있는지를 나타내는 지표이다.
대표적으로 다음 두 가지를 많이 사용한다.
- 분산(varience)
- 표준편차(standard deviation)

```python
# 분산 : 데이터가 평균에서 얼마나 떨어져 있는지 나타내는 값
# 값이 클수록 데이터의 퍼짐 정도가 크다는 의미
# (값 - 평균)^2의 평균
print(f'분산 : {np.mean((scores - np.mean(scores))**2)}')
print(f'분산 : {np.var(scores)}')
```

```python
# 표준 편차 : 분산의 제곱근
# 데이터가 평균에서 멀마나 퍼져 있는지를 직관적으로 나타냄
print(f'표준편차 : {np.sqrt(np.var(scores))}')
print(f'표준편차 : {np.std(scores)}')
```

## 모집단 VS 표본
분산과 표준편차는 계산 방식에 따라 두 가지가 있다.
- 모집단 : 던체 데이터를 알고 있을 때
- 표본: 일부 데이터로 전체를 추정할 때

Numpy에서는 ddof 옵션으로 구분한다. (분산 계산 시 나누는 값을 조정)
- ddof = 0 : 모집단(default)
- ddof = 1 : 표본

```python
# 모집단 기준 계산
print(f'모집단 분산 : {np.var(scores,ddof=0)}')
print(f'모집단 표준편차 : {np.std(scores,ddof=0)}')
# 표본 기준 계산 (분산 계산 시 나누는 값을 n-1로 처리하므로 값이 약간 크게 나온다.)
print(f'표본 분산 : {np.var(scores,ddof=1)}')
print(f'표본 표준편차 : {np.std(scores,ddof=1)}')
```

## 자주 상용하는 기본 통계 함수
Numpy는 데이터의 기본적인 통계 정보를 빠르게 확인 할 수 있는 함수를 제공한다.
이 함수들은 데이터 탐색(EDA)이나 간단한 통계 확인에서 자주 사용 된다.

```python
# 예시 데이터
values = np.array([12,25,8,45,33,17,29,41])

# 합계, 최대, 최소
print(np.sum(values))
print(np.min(values))
print(np.max(values))

# 최대값, 최소값의 위치
print(np.argmin(values))
print(np.argmax(values))

# 범위 : 데이터의 전체 퍼짐 정도 (max - min)
print(np.ptp(values))
```

## 난수 생성 (random data)
Numpy는 다양한 확률 분포 기반의 난수 생성 함수를 제공한다.
대표적으로 다음 분포를 많이 사용한다.
- 정규분포(normal distribution)
- 균등분포(uniform distribution)

데이터 시뮬레이션이나 테스트 데이터 생성에 사용된다.

### 정규 분포
평균을 중심으로 좌우 대칭인 종 모양 분포이다.
Numpy에서는 np.random.normal() 함수로 생성한다.

```python
# random.normal(loc,scale,size)
# loc : 평균값, scale : 표준편차, size : 개수
arr = np.random.normal(0,1,100)
arr = np.random.normal(0,1,(10,10))
arr
```

### 표준 정규 분포
np.random.randn()은 평균 0, 표준편차 1의 정규분포 데이터를 생성한다.

```python
arr = np.random.randn(100)
arr = np.random.randn(5,20)
arr
```

## 균등 분포
모든 값이 동일한 확률로 발생하는 분포이다.
np.random.randint() : 지정한 범위에서 정수 난수를 생성한다.

```python
# random.randint(low, high, size)
# low~high 범위의 size 개수 균등 분포 반환
arr = np.random.randint(1,100,1000)
arr
```

### np.random.rand() : 0~1 사이의 균등분포 실수 난수를 생성한다.

```python
arr = np.random.rand(10,10)
arr
```

## seed
난수 생성의 재현성을 위해 seed 값을 고정할 수 있다.

```python
np.random.seed(0)
np.random.rand(2,3)
```

## 샘플링
np.random.choice() 는 기존 데이터에서 무작위로 값을 선택할때 사용한다.
옵션
- replace = True : 복원 추출
- replace = False : 비복원 추출

```python
arr = np.array([10,3,5,7,8,9])
np.random.choice(arr)                           # 1개 추출
np.random.choice(arr,size=3)                    # 추출 개수 설정
np.random.choice(arr,size=3,replace=False)      # 중복 불허 비복원 추출
```

```python
# 확률 지정
probs = [0.8,0.05,0.05,0.05,0.025,0.025]
np.random.choice(arr,p=probs)
```
