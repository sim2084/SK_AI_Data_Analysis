## 0. 시작 데이터

```python
import numpy as np

# 3개 지점, 6개 상품의 재고 수량
stock = np.array([
    [12,  5, 18,  7, 10, 15],
    [ 8, 14, 11,  9,  6, 13],
    [16, 10,  9, 12, 14,  8]
])

# 상품 코드
product_ids = np.array([101, 102, 103, 104, 105, 106])

stock
```

## 1. 정렬
1) product_ids를 오름차순 정렬하시오.
2) 첫 번째 지점 재고를 오름차순 정렬하시오.
3) stock를 행 기준으로 정렬하시오.

```python
# 여기에 작성
print(np.sort(product_ids))
print(np.sort(stock[0]))
print(np.sort(stock,axis=1))
```

## 2. 역순 정렬
두 번째 지점 재고를 내림차순으로 정렬하시오.

```python
# 여기에 작성
print(np.sort(stock[1])[::-1])
```

## 3. argsort
세 번째 지점 재고를 많이 가진 상품 순서의 인덱스를 구하시오.
그리고 그 인덱스를 이용해 상품 코드도 같은 순서로 정렬하시오.

```python
# 여기에 작성
stock_indies = np.argsort(stock[2])[::-1]
print(stock_indies)

print(product_ids[stock_indies])
```

## 4. insert
첫 번째 지점 재고 앞에 신상품 재고 20을 삽입한 1차원 배열을 만드시오.

```python
# 여기에 작성
print(np.insert(stock[0],0,20))
```

## 5. delete
1) stock에서 마지막 상품 열을 삭제하시오.
2) stock에서 두 번째 지점 행을 삭제하시오.

```python
# 여기에 작성
print(np.delete(stock,-1,axis=1))
print(np.delete(stock,1,axis=0))
```

## 6. append / concatenate
아래 online_stock을 stock 뒤에 세로로 붙인 result_append를 만드시오.

```python
online_stock = np.array([[20, 18, 15, 10, 12, 17]])

# 여기에 작성
result_append = np.append(stock,online_stock,axis=0)
print(result_append)
```

## 7. vstack / hstack
1) stock 아래에 warehouse_stock을 세로로 붙이시오.
2) stock 오른쪽에 event_stock을 가로로 붙이시오.

```python
warehouse_stock = np.array([[30, 25, 28, 20, 26, 22]])
event_stock = np.array([[1], [2], [3]])

# 여기에 작성
print(np.vstack((stock,warehouse_stock)))
print(np.hstack((stock,event_stock)))
```

## 8. split
product_ids를 3개씩 나누시오.
그리고 stock를 열 기준으로 2개 블록으로 나누시오.

```python
# 여기에 작성
print(np.split(product_ids,2))
print(np.split(stock,2,axis=1))
```

## 9. reshape
1) 0부터 23까지의 값을 가지는 배열 arr를 만드시오.
2) arr를 4 x 6으로 바꾸시오.
3) arr를 2 x 3 x 4로 바꾸시오.

```python
# 여기에 작성
arr = np.arange(0,24)
print(arr.reshape(4,6))
print(arr.reshape(2,3,4))
```

## 10. -1 활용 reshape
arr를 이용해
1) 6행으로 reshape
2) 마지막 차원을 2로 갖도록 reshape
를 수행하시오.

```python
# 여기에 작성
print(arr.reshape(6,-1))
print(arr.reshape(-1,2))
```

## 11. ravel / flatten
matrix를 1차원으로 펼쳐서
1) ravel 결과
2) flatten 결과
를 각각 확인하시오.

```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

# 여기에 작성
print(matrix.ravel())
print(matrix.flatten())
```

## 12. expand_dims / squeeze
1) product_ids에 axis=0 축을 추가하시오.
2) product_ids에 axis=1 축을 추가하시오.
3) 추가한 축을 다시 제거하시오.

```python
# 여기에 작성
print(np.expand_dims(product_ids,axis=0))
print(np.expand_dims(product_ids,axis=1))
print(np.expand_dims(product_ids,axis=1).squeeze())
```

## 13. transpose
stock의 행과 열을 바꾼 배열을 만드시오.

```python
# 여기에 작성
print(stock.T)
```

## 14. 종합 문제
아래 daily_stock은 2개 지점의 3일치, 4개 상품 재고이다.
shape는 (2, 3, 4) 이다.
1) 하루 단위로 보기 쉽게 6 x 4 형태로 reshape 하시오.
2) 각 행의 재고 합계를 구하시오.
3) 재고 합계가 작은 순서대로 행 인덱스를 구하시오.
4) 가장 재고 합계가 작은 하루의 데이터만 선택하시오.

```python
daily_stock = np.array([
    [[10, 12,  8,  9],
     [11, 13,  7, 10],
     [ 9, 14,  8, 11]],

    [[15, 10, 12, 13],
     [14, 11, 10, 12],
     [13,  9, 11, 10]]
])

# 여기에 작성
day_stock = daily_stock.reshape(1,6,4).squeeze()
print(day_stock)

print(day_stock.sum(axis=1))
print(np.argsort(day_stock.sum(axis=1)))
print(day_stock[np.argsort(day_stock.sum(axis=1)).min()])
```
