# Indexing | Slicing
- 배열에서 원하는 값을 선택하는 방법이다.
- Numpy에서는 특정 위치의 값 조회, 특정 구간의 데이터 추출, 조건 기반 데이터 선택 등을 자주 사용한다.

## Indexing
- 배열에서 특정 위치(인덱스) 값을 하나 선택하는 방법이다.
- 파이썬의 list와 동일하게 0부터 시작하며 -1은 마지막 요소를 의미한다.

```python
import numpy as np
```

```python
# 1차원
arr = np.arange(1,11)
print(arr)
print(arr[0],arr[1])
print(arr[-1],arr[-2])
# print(arr[100])       # 인덱스 범위 벗어나면 오류
```

```python
# 2차원
arr1 = np.arange(1,11)
print(arr1)
arr1 = arr1.reshape(2,5)
print(arr1)

print(arr1[0][2],arr1[0,2])
print(arr1[1][3],arr1[1,3])
print(arr1[-1][-2],arr1[-1,-2])
```

## Slicing
- 배열에서 특정 구간의 데이터를 잘라서 가져오는 방법이다.
- start : stop : step 형태를 사용하며, 위치는 포함되지 않는다.

```python
# ndarray[start:stop:step]
arr1 = np.arange(1,11)
print(arr1)
print(arr1[2:7])
print(arr1[2:])
print(arr1[:7])
print(arr1[:-1])
print(arr1[:])
print(arr1[:-2:2])
print(arr1[::-1])
```

```python
# 2차원 배열 슬라이싱
arr = np.arange(1,13).reshape(3,4)
print(arr[:,:],end='\n------------------\n')     # 모든 행, 모든열
print(arr[0,:],end='\n------------------\n')
print(arr[:,0],end='\n------------------\n')
print(arr[:,0:2],end='\n------------------\n')
print(arr[1:,2:],end='\n------------------\n')
```

## Fancy indexing
- 여러 개의 특정 위치를 한 번에 선택하는 방법이다.
- 인텍스 리스트를 사용하여 원하는 요소만 추출할 수 있다.

```python
# 1차원
arr = np.linspace(5,50,10, dtype=int)
print('arr: ',arr)
print('arr[1,2,8]: ', arr[[1,2,8]] )

indexes = [1,2,8]
print('arr[indexes]: ',arr[indexes])
```

```python
# 2차원
arr = np.linspace(5,100,20,dtype=int).reshape(4,5)
print('arr: ',arr)

print('arr[0,1]: ',arr[0,1])            # 좌표 근접 arr[0][1]
print('arr[[0,1]]: ',arr[[0,1]])        # 행 선택 - 0, 1행
print('arr[:,[0,2]]: ',arr[:,[0,2]])    # 열 선택 - 0, 2열

# 행과 열이 모두 리스트면 좌표 선택 - 0행 1열, 1행 3열
print('arr[[0,1],[1,3]]: ',arr[[0,1],[1,3]])

print('arr[[0,1],1:4]: ',arr[[0,1],1:4])
print('arr[[1,3],3:]: ',arr[[1,3],3:])
print('arr[[1,3],-2:]: ',arr[[1,3],-2:])
```

## Boolean Indexing
- 조건식의 결과(True/False)를 이용해서 조건을 만족하는 데이터만 선택하는 방법
- 데이터 분석에서 필터링 할 때 자주 사용 된다.

```python
arr = np.array([1,2,3,4,5])
print(arr[[1,3]])
print(arr[[False,True,False,True,False]])

print(arr % 2 == 0)
print(arr[arr % 2 == 0])

print(arr>3)
print(arr[arr>3])
```
