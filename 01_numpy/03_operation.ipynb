# ndarray 연산

## element-wise 연산
- 기본적으로 두 배열은 동일한 shape여야한다.
- 같은 위치(인덱스)의 요소끼리 연산이 수행 된다.
- 결과 배열 또한 동일한 shape를 가진다.

```python
import numpy as np

arr1 = np.arange(1,10).reshape(3,3)
print(arr1)

arr2 = np.full_like(arr1,2)
print(arr2)
```

```python
print(arr1 + arr2)   # 덧셈
print(arr1 - arr2)   # 뺄셈
print(arr1 * arr2)   # 곱셈(요소별 곱)
print(arr1 / arr2)   # 나눗셈
print(arr1 // arr2)  # 몫
print(arr1 % arr2)   # 나머지
print(arr1 ** arr2)  # 거듭제곱
```

## broadcasting 연산 방식
- 두 배열의 shape이 다르면 broadcasting 규칙에 따라 연산이 수행 된다
- 차원이 적은 배열을 자동으로 확장하여 연산이 가능하도록 한다.

### 규칙
- 뒤쪽(마지막 차원)부터 shape를 비교
- 두 차원이 같거나, 둘중 하나가 1이면 연산 가능
- 둘 다 다르고 둘 다 1이 아니면 broadcasting 불가


```python
# 브로드캐스트 연산 테스트
arr1 = np.array([[1, 2, 3], [4,5,6]])
arr2 = np.full((3,),2)

print(arr1,arr1.shape)
print(arr2,arr2.shape)

# (2,3) 배열과 (3,) 배열의 연산
# (3,) -> (1,3) -> (2,3)
print(arr1+arr2)
```

```python
arr3 = np.array([1,2])
print(arr3, arr3.shape)
# ValueError: operands could not be broadcast together with shapes (2,3) (2,)
# print(arr1 + arr3)
```

```python
# arr1과 다양한 연산
print(arr1 + 10)        # (2,3) + () 스칼라도 브로드캐스팅 적용
print(arr1 + [10])      # (2,3) + (1,) 브로드캐스팅 적용
print(arr1 + [[10]])    # (2,3) + (1,1) 브로드캐스팅 적용
```

## 행렬곱셈 Matrix Multiplication
점곱연산 Dot Product (내적)

두 행렬 A B의 곱셈은 첫번째 행렬 A의 행과 두번째 행렬 B의 열간의 곱셈을 수행한다.

조건
- 첫번째 행렬 A의 열의 수와 두번째 행렬 B의 행의 수가 같아야 한다.
- 연산결과는 (첫번째 행렬 A의 행수, 두번째 행렬 B의 열수) 이다.

- **수식**: 행렬 $ A $가 $ m \times n $ 크기이고, 행렬 $ B $가 $ n \times p $ 크기일 때, 행렬 곱 $ A \times B $의 결과는 $ m \times p $ 크기의 행렬이다. 곱셈 과정은 다음과 같다:
  $$
  (A \times B)_{ij} = \sum_{k=1}^{n} A_{ik} \times B_{kj}
  $$
  즉, $ A $의 i번째 행과 $ B $의 j번째 열에 해당하는 원소들을 곱하고, 그 결과를 더하여 새로운 행렬의 $ (i, j) $ 위치에 배치한다.

- **예시**:
  $$
  A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}, \quad B = \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix}
  $$
  $$
  A \times B = \begin{pmatrix} 1 \cdot 5 + 2 \cdot 7 & 1 \cdot 6 + 2 \cdot 8 \\ 3 \cdot 5 + 4 \cdot 7 & 3 \cdot 6 + 4 \cdot 8 \end{pmatrix} = \begin{pmatrix} 19 & 22 \\ 43 & 50 \end{pmatrix}
  $$

- **결과**: 새로운 **행렬**.

NumPy에서는 다음 세 가지 방법으로 행렬곱을 수행할 수 있다.
- @ 연산자
- np.matmul()
- np.dot() -> 2차원은 동일하나 다차원으로 올라가면 다르게 동작할 수 있으므로 사용 X

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])

print(a @ b)
print(np.matmul(a,b))
print(np.dot(a,b))
```

```python
c = np.array([[1,2]])       # (1,2)
d = np.array([[3],[4]])     # (2,1)
print(c.shape,d.shape)
x = c @ d
print(x.shape, x)
```

```python
c = np.array([[1,2,3],[4,5,6]])             # (2,3)
d = np.array([[7,8],[9,10],[11,12]])        # (3,2)
print(c.shape,d.shape)
x = c @ d
print(x.shape, x)                           # (2,2)
```

```python
c = np.array([[1,2,3],[4,5,6]])             # (2,3)
d = np.array([[7,8,9],[10, 11,12]])        # (2,3)
print(c.shape,d.shape)
# x = c @ d                                3과 2는 행렬 곱 불가 ValueError 발생
# print(x.shape, x)   
```

## 기타 연산
Numpy는 배열 전체에 대해 다양한 수학 함수를 제공한다.

```python
# 절대값
arr1 = np.array([[1,-2],[-3,4]])
print(np.abs(arr1))
```

```python
# 반올림
arr2 = np.array([-1.234,1.234,2.345,3.456,4.567])
print(np.round(arr2))       #반올림
print(np.ceil(arr2))        #올림
print(np.floor(arr2))       #내림
print(np.trunc(arr2))       #절삭
```

### axis 속성
연산을 수행할 축을 지정한다.

- axis = 0 : 행 방향으로 연산 (열 기준 연산)
- axis = 1 : 열 방향으로 연산 (행 기준 연산)
- axis = None : 배열을 1차원으로 펼친 후 연산

- axis는 "어느 방향으로 줄일 것이가"를 의미

```python
# min
arr = np.array([[1,2,3],[4,5,6]])
print(np.min(arr))
print(arr.min())    # axis=None 이 기본 값으로 배열을 1차원으로 펼치고 최솟값을 구함

print(arr.min(axis=0))
print(arr.min(axis=1))
```

```python
# sum
print(arr.sum())
print(arr.sum(axis=0))
print(arr.sum(axis=1))
```

```python
# mean
print(arr.mean())
print(arr.mean(axis=0))
print(arr.mean(axis=1))
```
