# EDA (Exploratory Data Analysis)

## 01. 데이터 수집 및 로딩
- CSV, Excel, DB, API 등에서 데이터를 불러오고, DataFrame 생성

```python
import pandas as pd
df = pd.read_csv('data.csv')
```

---

## 02. 데이터 구조 및 변수 이해
- 데이터 목적 파악, Feature/Target 구분
- 변수 타입 확인 (수치형, 범주형 등)
- 크기, 컬럼명, 샘플 확인

```python
print(df.shape)
print(df.dtypes)
print(df.head())
print(df.info())
```

---

## 03. 데이터 요약 및 기술 통계
- describe로 기본 통계량 확인
- 범주형은 value_counts로 분포 확인

```python
print(df.describe())
print(df['gender'].value_counts())
```

---

## 04. 결측치 및 이상치 탐색
- isnull로 결측치 개수 확인
- boxplot 등으로 이상치 탐지

```python
print(df.isnull().sum())
df.boxplot(figsize=(10, 5))
```

---

## 05. 변수 분포 시각화
- 히스토그램, 박스플롯 등으로 개별 변수의 분포 확인

```python
import seaborn as sns
sns.histplot(df['age'])
sns.boxplot(x='gender', y='age', data=df)
```

---

## 06. 변수 간 관계 시각화
- 산점도, pairplot 등으로 변수 간 상관관계 확인

```python
sns.scatterplot(x='age', y='income', data=df)
sns.pairplot(df[['age', 'income', 'score']])
```

---

## 07. 상관관계 및 교차분석
- 수치형: 상관계수
- 범주형: 교차표

```python
print(df.corr())
pd.crosstab(df['gender'], df['target'])
```

---

## 08. 데이터 전처리
- 결측치 제거/대체, 타입 변환, 불필요한 컬럼 제거

```python
df = df.dropna()
df['age'] = df['age'].astype(int)
```

---

## 09. Feature Engineering
- 새로운 변수 생성, 그룹화, 로그변환, 스케일링 등

```python
df['age_group'] = pd.cut(df['age'], bins=[0,20,40,60,100], labels=['Teen','Adult','Middle','Senior'])
```

---

## 10. 최종 요약 및 인사이트 도출
- 주요 변수 통계 요약
- 변수 간 영향력 확인
- 분석/모델링에 필요한 방향 설정

```python
# 예시: 그룹별 평균 요약
df.groupby('gender')['income'].mean()
```

# 머신러닝 생명주기 (ML Lifecycle)

머신러닝 프로젝트는 단순한 모델 학습을 넘어서 전체 시스템 흐름을 고려해야 한다.

| 단계               | 설명                                    |
| ---------------- | ------------------------------------- |
| 1. 문제 정의         | 예측 또는 분류할 문제를 정의 (비즈니스 목표 포함)         |
| 2. 데이터 수집        | 데이터베이스, API, 로그 등에서 데이터 확보            |
| 3. 데이터 전처리 및 EDA | 결측치/이상치 처리, EDA 수행, 피처 엔지니어링          |
| 4. 데이터 분할        | 학습/검증/테스트 데이터 분할 (train/test split)   |
| 5. 모델 선택 및 학습    | 알고리즘 선정 (예: RandomForest, SVM 등) 및 학습 |
| 6. 모델 평가         | 정확도, 정밀도, F1-score, ROC-AUC 등으로 성능 평가 |
| 7. 모델 튜닝         | 하이퍼파라미터 조정 및 성능 향상                    |
| 8. 배포 및 서비스      | API 형태로 배포하거나 시스템에 내장                 |
| 9. 모니터링 및 유지보수   | 성능 저하 감지, 모델 업데이트 수행                  |


# 정리 비교

| 구분    | EDA               | 머신러닝 생명주기          |
| ----- | ----------------- | ------------------ |
| 역할    | 데이터 이해 및 전처리 준비   | 모델 개발과 배포 전 과정을 관리 |
| 시점    | 모델링 이전            | 전체 프로젝트 기간         |
| 핵심 활동 | 통계 분석, 시각화, 변수 탐색 | 문제 정의 ~ 배포 및 유지보수  |
