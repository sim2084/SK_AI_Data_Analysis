## 1. 가장 기본적인 그래프
`x`와 `y` 값을 준비한 뒤 `plt.plot()`으로 선 그래프를 그린다. 마지막에는 `plt.show()`로 화면에 출력한다.

```python
# x축, y축 데이터
X = [1, 2, 3]
Y = [2, 4, 6]

plt.plot(X, Y)  # 선 그래프 그리기
plt.title("Hello Matplotlib")  # 그래프 제목 설정

plt.show()  # 그래프 화면에 출력
```

## 2. Figure와 Axes
Figure는 전체 도화지이고, Axes는 실제 그래프가 그려지는 영역이다. 하나의 Figure 안에 여러 Axes가 들어갈 수 있다.

- Figure: 전체 그림을 그리는 틀 (캔버스)
- Axes: 그래프가 실제로 그려지는 영역 (데이터와 축을 포함)

**주요명령어**
- 새 figure 생성하기 `plt.figure()`
- 새 axes 생성하기 `plt.axes()`
- 현재 axes 가져오기(get_current_axes) `ax = plt.gca()`
- figure와 그에 속한 axes 동시에 생성하기 `fig, ax = plt.subplots()`

![](images/anatomy_of_a_figure.png)


- Figure: 전체 그래프 캔버스 (이미지 한 장)
- Axes: 실제 그래프가 그려지는 영역 (그래프 하나)
- Axis (X axis / Y axis): x축, y축
- Spines: 그래프 영역의 테두리 선

- Title: 그래프 전체 제목
- X axis label: x축 이름
- Y axis label: y축 이름

- Major tick: 큰 눈금
- Minor tick: 작은 눈금
- Major tick label: 큰 눈금에 표시되는 값
- Minor tick label: 작은 눈금에 표시되는 값

- Grid: 격자선 (데이터 위치 파악용)

- Line (line plot): 선 그래프
- Markers (scatter plot): 점 (산점도 데이터)

- Legend: 범례 (그래프 요소 설명)

```python
# Figure 생성 (전체 캔버스 설정)
plt.figure(figsize=(8, 6), facecolor='yellow')  # 크기(가로,세로) 인치, 배경색

# 현재 Axes 가져오기 (그래프가 실제로 그려질 영역)
ax = plt.gca()

# Axes에 그래프 그리기
ax.plot(X, Y)

# 그래프 제목 설정
plt.title("Figure - Axes")

# 화면에 출력
plt.show()
```

figsize=(6, 4): 인치(inch) 단위, DPI (dots per inch) 에 따라 결정. 즉 대략 600 x 400 픽셀 느낌.
- (6,4) : 기본 그래프
- (8,6) : 조금 넓은 발표용
- (12,8) : 보고서/슬라이드용

## 3. 여러 개의 Axes 배치하기
`plt.subplots()`를 사용하면 여러 그래프를 한 번에 비교할 수 있다. 여기서는 2행 3열 구조를 만든다.

```python
# 여러 개의 Axes 생성 (2행 3열 구조)
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(8, 6))

print(ax.shape)  # axes 배열 형태 확인 (2, 3)

# 첫 번째 행 (line plot)
ax[0, 0].plot(X, Y)  # 1행 1열
ax[0, 1].plot(X, Y)  # 1행 2열
ax[0, 2].plot(X, Y)  # 1행 3열

# 두 번째 행 (scatter plot)
ax[1, 0].scatter(X, Y)  # 2행 1열
ax[1, 1].scatter(X, Y)  # 2행 2열
ax[1, 2].scatter(X, Y)  # 2행 3열

plt.show()  # 전체 figure 출력
```

```python
# TODO
# 아래 조건을 만족하도록 2행 2열 Axes를 구성해보세요.
# 1) figsize를 (8, 6)으로 설정하기
# 2) 첫 번째 행은 선 그래프(Y), 두 번째 행은 막대 그래프(Y)로 구성하기
# 3) 왼쪽은 기본 그래프, 오른쪽은 색상을 변경한 그래프로 그리기
# 4) 각 Axes의 제목을 'line1', 'line2', 'bar1', 'bar2'로 설정하기 (ax.set_title 사용)
# 5) 모든 Axes에 x축, y축 이름을 설정하기
fig, ax = plt.subplots(2,2,figsize = (8,6))

# 1행 - 선그래프
ax[0,0].plot(X, Y)
ax[0,0].set_title('line1')
ax[0,0].set_xlabel('X')
ax[0,0].set_ylabel('Y',rotation = 0)
ax[0,1].plot(X, Y, color = 'green')
ax[0,1].set_title('line2')
ax[0,1].set_xlabel('X')
ax[0,1].set_ylabel('Y',rotation = 0)

# 2행 - 막대그래프
ax[1,0].bar(X, Y)
ax[1,0].set_title('bar1')
ax[1,0].set_xlabel('X')
ax[1,0].set_ylabel('Y',rotation = 0)
ax[1,1].bar(X, Y, color = 'orange')
ax[1,1].set_title('bar2')
ax[1,1].set_xlabel('X')
ax[1,1].set_ylabel('Y',rotation = 0)

plt.tight_layout()
plt.show()
```

## 4. plot 주요 옵션
색상(`color`), 선 모양(`linestyle`), 마커(`marker`), 범례(`label`, `legend`)를 조절해보자.

```python
# 선 그래프 + 스타일 옵션 설정
plt.plot(
    X, Y,
    color='blue',          # 색상: 'blue', 'red', 'green', 'black', 'orange', 'purple' ...
    linestyle='dashed',    # 선 스타일: 'solid', 'dashed', 'dashdot', 'dotted'
    linewidth=2,           # 선 두께: 1, 2, 3 ...
    marker='o',            # 마커: 'o'(원), 's'(사각형), '^'(삼각형), 'x', '*'
    markersize=10,         # 마커 크기: 숫자 (5~15 정도 추천)
    label='my line plot'   # 범례 이름 (문자열 자유롭게)
)

# 축 이름(label)
plt.xlabel("X")  # x축 이름 (문자열 변경 가능)
plt.ylabel("Y")  # y축 이름 (문자열 변경 가능)

# 눈금(ticks) 설정
plt.xticks(ticks=[1, 2, 3])  
# 리스트 값 변경 가능: [0, 1, 2, 3] / range(0, 5)

plt.yticks(ticks=range(0, 7), rotation=45)  
# ticks: range(0, 10), [0, 2, 4, 6] 등
# rotation: 0, 45, 90

# 축 범위 제한 (줌인/줌아웃 효과)
plt.xlim(2, 3)   
# 범위 변경: (0, 3), (1, 4)

plt.ylim(0, 10)  
# 범위 변경: (0, 6), (2, 8)

# 범례(legend) 표시
plt.legend(loc='lower center')  
# 위치: 'best', 'upper right', 'upper left', 'lower right', 'lower left', 'center'

# 그래프 출력
plt.show()
```

## 5. 여러 개의 plot 한 화면에 그리기
한 Axes에 선을 여러 개 올리면 서로 다른 추이를 직접 비교할 수 있다.

```python
# 여러 개의 plot을 한 화면에 그리기
import numpy as np

# 데이터 생성
X = np.arange(1, 10)  # 1~9까지 값 생성
Y = X * 2             # 선형 데이터 (2배)
Z = X * 4             # 선형 데이터 (4배)

# 선 그래프 추가
plt.plot(X, Y, color='red', label='X-Y')  
# color: 'red', 'blue', 'green' ...
# label: 범례 이름

# 막대 그래프 추가 (같은 Axes 위에 겹쳐서 그려짐)
plt.bar(X, Z, color='blue', label='X-Z')  
# bar: 막대 그래프
# color 변경 가능

# 범례 표시 (그래프 구분)
plt.legend()

# 그래프 출력
plt.show()
```

```python
# TODO
# 아래 조건을 만족하도록 한 화면에 그래프를 다시 그려보세요.
# 1) X는 1~7까지 생성하기
# 2) Y는 "방문자 수", Z는 "구매 수"로 데이터를 직접 생성하기 (값은 자유롭게 설정)
# 3) 방문자 수(Y)는 선 그래프로, 구매 수(Z)는 막대 그래프로 표현하기
# 4) 선 그래프는 color='purple', marker='o'로 설정하기
# 5) 막대 그래프는 color='orange'로 설정하기
# 6) 그래프 제목을 '방문자 vs 구매'로 설정하기
# 7) 범례를 추가하기
# 8) y축 이름을 '수량'으로 설정하기
# 9) grid를 표시하기

X = np.arange(1,8)
Y = [50,60,55,45,75,80,65]
Z = [10,8,7,6,12,15,13]

plt.plot(X,Y,color='purple',marker='o',label = '방문자 수')
plt.bar(X,Z,color='orange',label = '구매 수')
plt.title('방문자 vs 구매')
plt.legend()
plt.ylabel('수량')
plt.grid()

plt.show()
```

## 정리
- `plt.plot()`은 가장 기본적인 선 그래프 함수이다.
- Figure는 도화지, Axes는 실제 그래프 영역이다.
- `subplots()`로 여러 그래프를 배치할 수 있다.
- Matplotlib은 세부 제어가 강한 대신 직접 설정할 것이 많다.
- 이어서 차트 유형별로 언제 어떤 그래프를 써야 하는지 이어서 확인한다.
